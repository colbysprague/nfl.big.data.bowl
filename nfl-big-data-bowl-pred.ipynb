```python
from fastai.tabular.all import *
from sklearn.ensemble import RandomForestRegressor
```

```python
"""
    Returns a dls given some list of dependent variable columns
"""
def get_dls_for_dep_vars(dep_vars: list[str]):
    cat_names = ['play_direction',
                 'player_name',
                 'player_position',
                 'player_side',
                 'player_role',
                 'player_height',
                ]
    
    cont_names = ['absolute_yardline_number',
                  'player_weight',
                  's',
                  'a',
                  'dir',
                  'o',
                  'num_frames_output', 
                  'ball_land_x', 
                  'ball_land_y'
                 ]
    
    splits = RandomSplitter(valid_pct=0.2)(range_of(df))
    
    return TabularPandas(
        df, 
        procs=[Categorify, FillMissing, Normalize],
        cat_names=cat_names,
        cont_names=cont_names,
        y_names=dep_vars,
        splits=splits
    )
```

```python
tp_x = get_dls_for_dep_vars([X])
tp_y = get_dls_for_dep_vars([Y])

def train_dls(tab_pand): 
    
    dls = tab_pand.dataloaders(bs=64)
    learn = tabular_learner(dls, metrics=rmse, layers=[200, 100])
    
    learn.fit_one_cycle(5)

    return learn

learn_x = train_dls(tp_x)
learn_y = train_dls(tp_y)
```

```python
def evaluate_model(learn, test_df):
    """
    Evaluate model on test data using existing dataloader structure
    
    Args:
        learn: Trained fastai learner
        test_df: Test dataframe with target column
    
    Returns:
        dict: Dictionary of metrics
    """
    import numpy as np
    from sklearn.metrics import mean_absolute_error, mean_squared_error
    
    # Use the existing dataloader with new data
    test_dl = learn.dls.test_dl(test_df)
    
    # Get predictions and targets
    preds, targets = learn.get_preds(dl=test_dl)
    
    # Convert to numpy and flatten (since it's single output)
    preds_np = preds.numpy().flatten()
    targets_np = targets.numpy().flatten()
    
    # Calculate metrics
    results = {}
    results['mae'] = mean_absolute_error(targets_np, preds_np)
    results['rmse'] = np.sqrt(mean_squared_error(targets_np, preds_np))
    results['mse'] = mean_squared_error(targets_np, preds_np)
    
    # Additional statistics
    errors = preds_np - targets_np
    results['mean_error'] = np.mean(errors)
    results['median_error'] = np.median(errors)
    results['std_error'] = np.std(errors)
    
    # Print results
    print("="*50)
    print("MODEL EVALUATION RESULTS")
    print("="*50)
    print(f"MAE:                        {results['mae']:.4f}")
    print(f"RMSE:                       {results['rmse']:.4f}")
    print(f"MSE:                        {results['mse']:.4f}")
    print(f"-"*50)
    print(f"Mean Error:                 {results['mean_error']:.4f}")
    print(f"Median Error:               {results['median_error']:.4f}")
    print(f"Std Error:                  {results['std_error']:.4f}")
    print("="*50)
    
    return results
```

```python
def get_predictions_df(learn, test_df, pred_col_name='prediction'):
    """
    Get predictions and append them to the test dataframe
    
    Args:
        learn: Trained fastai learner
        test_df: Test dataframe with target column
        pred_col_name: Name for the prediction column (default: 'prediction')
    
    Returns:
        DataFrame: Copy of test_df with predictions appended
    """
    import numpy as np
    
    # Use the existing dataloader with new data
    test_dl = learn.dls.test_dl(test_df)
    
    # Get predictions
    preds, _ = learn.get_preds(dl=test_dl)
    
    # Convert to numpy and flatten (since it's single output)
    preds_np = preds.numpy().flatten()
    
    # Create a copy of the dataframe and add predictions
    result_df = test_df.copy()
    result_df[pred_col_name] = preds_np
    
    return result_df
```
